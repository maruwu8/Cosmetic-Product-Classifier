```python
import pandas as pd
```

```python
data = pd.read_csv("products.csv")
display(data)
```

```python
columns = data.columns
titles = [column.title() for column in columns]
print(titles)
```

```python
display(data.describe())
```

```python
data = data.dropna(subset=["Product Name"])
data = data.drop_duplicates()
```

```python
display(data.describe())
display(data)
```

```python
total_products = len(data)
print("Total Number of Products:", total_products)
```

```python
products_per_category = data['Category'].value_counts()
print("Number of Products per Category:\n", products_per_category)
```

```python
unique_brands = data['Brand'].nunique()
print("Number of Brands:", unique_brands)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

data['Price'] = pd.to_numeric(data['Price'], errors='coerce')

average_price_per_brand = data.groupby('Brand')['Price'].apply(lambda x: x.dropna().mean())
average_price_per_brand = average_price_per_brand.dropna().sort_values(ascending=False)

# Distribution of Prices
plt.figure(figsize=(10, 6))
sns.histplot(data['Price'].dropna(), bins=20, kde=True)
plt.title('Distribution of Product Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()

print("Average Prices per Brand (Descending Order):\n", average_price_per_brand)
```

```python
# Visualization of Products per Category
plt.figure(figsize=(10, 6))
sns.barplot(x=products_per_category.index, y=products_per_category.values)
plt.title('Number of Products per Category')
plt.xticks(rotation=45)
plt.xlabel('Category')
plt.ylabel('Number of Products')
plt.show()

# Visualization of Price Analysis per Category
price_analysis = data.groupby('Category')['Price'].agg(['count', 'mean', 'min', 'max'])
plt.figure(figsize=(10, 6))
sns.barplot(x=price_analysis.index, y=price_analysis['mean'])
#sns.barplot(x=price_analysis.index, y=price_analysis['max'])
plt.title('Average Price per Category')
plt.xticks(rotation=45)
plt.xlabel('Category')
plt.ylabel('Average Price')
plt.show()
```
